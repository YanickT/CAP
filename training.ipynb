# Prepare Datasets

```python
import os
import random

path = "/content/drive/MyDrive/RIAFML"
files = [file for file in os.listdir(path) if not os.path.isdir(f"{path}/{file}")]

# create empty folders
os.mkdir(f"{path}/test")
os.mkdir(f"{path}/train")

# randomly choose 10% of the data as training data
random.shuffle(files)
test_files = files[::10]
train_files = [file for file in files if not (file in test_files)]

# sort files to correct folders
for file in test_files:
  os.rename(f"{path}/{file}", f"{path}/test/{file}")

for file in train_files:
  os.rename(f"{path}/{file}", f"{path}/train/{file}")
```

# Start Training

```python
import torch
import random
import os

# check hardware 
torch.cuda.empty_cache()
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print(f"Device: {device}")
```

```python
from drive.MyDrive.CAP.network import SpinNetwork, IncNetwork
from drive.MyDrive.CAP.datagen import DataGen 


TARGET = "inc"  # change to "a" to load targets for spin
path = "/content/drive/MyDrive/RIAFML"


train_files = [file for file in os.listdir(f"{path}/train") if file[-4:] == ".npz"]
random.shuffle(train_files)
train_loader = DataGen(train_files, TARGET, path=f"{path}/train", device=device, bs=32)

test_files = [file for file in os.listdir(f"{path}/test") if file[-4:] == ".npz"]
random.shuffle(test_files)
test_loader = DataGen(test_files, TARGET, path=f"{path}/test", device=device, bs=32)
print(f"Prepared {len(train_files)} train files with {len(test_files)} test files")
```

```python
import matplotlib.pyplot as plt

losses = []
net = IncNetwork(device=device)  # change to SpinNetwork to train the network for spin
```

```python
lr = 1e-3
loss = net.train(train_loader, test_loader, its=5, lr=lr)
losses.append((lr, loss))

plt.plot(loss)
plt.show()
```
